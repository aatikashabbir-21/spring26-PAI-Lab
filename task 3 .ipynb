```python
def water_jug_dfs(capacity1, capacity2, target):

    stack = [(0, 0, [])]
    visited = set()

    while stack:
        jug1, jug2, path = stack.pop()

        if (jug1, jug2) in visited:
            continue

        visited.add((jug1, jug2))
        print(f"Current State: ({jug1}, {jug2})")

        if jug1 == target or jug2 == target:
            print("\nGoal Reached!\n")
            for step in path:
                print(step)
            print(f"\nFinal State: ({jug1}, {jug2})")
            return

        new_states = []

        new_states.append((capacity1, jug2,
                           path + [f"Fill Jug1 -> ({capacity1}, {jug2})"]))

        new_states.append((jug1, capacity2,
                           path + [f"Fill Jug2 -> ({jug1}, {capacity2})"]))

        new_states.append((0, jug2,
                           path + [f"Empty Jug1 -> (0, {jug2})"]))

        new_states.append((jug1, 0,
                           path + [f"Empty Jug2 -> ({jug1}, 0)"]))

        transfer = min(jug1, capacity2 - jug2)
        new_states.append((jug1 - transfer, jug2 + transfer,
                           path + [f"Pour Jug1 -> Jug2 -> ({jug1 - transfer}, {jug2 + transfer})"]))

        transfer = min(jug2, capacity1 - jug1)
        new_states.append((jug1 + transfer, jug2 - transfer,
                           path + [f"Pour Jug2 -> Jug1 -> ({jug1 + transfer}, {jug2 - transfer})"]))

        for state in new_states:
            if (state[0], state[1]) not in visited:
                stack.append(state)

    print("No Solution Found.")


capacity1 = 4
capacity2 = 3
target = 2

print("Water Jug Problem using DFS\n")
water_jug_dfs(capacity1, capacity2, target)
```
